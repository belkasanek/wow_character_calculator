```python
import pandas as pd

from character import Character
```

```python
# create new instance of a class
s = Character(game_class='warlock', race='orc')
```

```python
# summary method return dict of stats
s.summary()
```

`search` is one of main method in class. It helps to search in item table for best gear. The method has one required argument `slot`. Other usefull arguments are `armor_type` (cloth, leather, mail, plate), `quality` (epic, rare, common etc),  `orderby` take a list of strings which is used to sort table

```python
temp = s.search(slot='head', armor_type='cloth', quality='rare', 
                orderby=['Increase Spell Dam', 'Increase Shadow Dam', 'stamina', 'Increased Critical Spell'])
temp[['id', 'name', 'Increase Spell Dam', 'Increase Shadow Dam', 'stamina', 'Increased Critical Spell']].head(10)
```

`wear_item` method requirs `id` of an item to put on it on character

```python
s.wear_item(ids=10041)
```

If you change your mind yon can remove item from your character by `remove_item` method. It requirs `slot` name of an item 

```python
s.remove_item(slot='head')
s.wear_item(ids=18727)
```

Armor type in `search` method for neck doesn't make sence. You can leave it empty or fill it with some valid armor type. It will be ignored. 

```python
temp = s.search(slot='neck', armor_type='cloth', quality='rare', 
                orderby=['Increase Spell Dam', 'Increase Shadow Dam', 'stamina', 'Increased Critical Spell'])
temp[['id', 'name', 'Increase Spell Dam', 'Increase Shadow Dam', 'stamina', 'Increased Critical Spell']].head(10)
```

You can transpose table and remove columns filtering

```python
temp = s.search(slot='shoulders', armor_type='cloth', quality='rare', 
                orderby=['Increase Spell Dam', 'Increase Shadow Dam', 'stamina', 'Increased Critical Spell'])
temp.head(10).T
```

`search` return pandas DataFrame, so you can use usual method to filter results further.

```python
temp.loc[(temp['stamina'] >= 15) & (temp['intellect'] >= 15) & 
         ((temp['Increase Spell Dam'] >= 10) | (temp['Increase Shadow Dam'] >= 10)),
        ['id', 'name', 'Increase Spell Dam', 'Increase Shadow Dam', 'stamina', 'Increased Critical Spell']]
```
